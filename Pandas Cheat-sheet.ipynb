## What is Pandas ?
- Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,
built on top of the Python programming language.
## Why Pandas ?
- To use the the flexiblity and power of python.
- The functionality of Pandas is built on top of NumPy and both libraries belong to the SciPy stack
- Pandas helps to clean the mess.
- Read CSV and Excel file data in chunk size
- Filter out unimportant columns to save memory
- To Change dtypes for columns.
- When we think of Python for Data Analysis NumPy and Pandas are to be involved.

## Reading Data in Pandas.
### Loading data into Pandas and in this syntax 'df ' stands for data-frame


```python
df = pd.read_csv('pokemon_data.csv')
print(df) # For printing the whole data-frame
```

```python
# To print the data on the specific location
print(df.iloc[0,3]) # It works same as we use to print an specific element of a matrix using[r,c]
```

### Sorting / Describing Data

```python
df.sort_values(['Type 1', 'HP'],ascending = [1,0])
df
```

### Changing the Data

```python
# Adding a new column and feeding the data in it by doing arthamatic operation between the columns.
df['Total'] = df['HP'] + df['Attack'] + df['Defense']+ df['Sp. Atk'] + df['Sp. Def'] +df['Speed']
df
```

```python
# Deleting a column.
df = df.drop(columns = ['Total'])
df
```

```python
# The same operation performed in the line no. 14 can also be performed in a more compact and efficient way.
df['Total'] = df.iloc[:, 4:10].sum(axis = 1)
# Here we can see that one new column has been added i.e. "Total"
df
```

```python
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]] # Here we have rearranged the columns as the 13th column is been shifted to 4th position  
df
```
